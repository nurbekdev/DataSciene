## `NaN` qiymatlarni tashlab yuboramiz

```python
df.dropna()
```

```python
len(df.dropna())
```

```python
df.dropna(axis=1) #Nan qiymatli Ustunlarni tashlab yuborish
```

```python
df.dropna(axis=1, thresh=48) # Chegara
```

```python
df.dropna(subset=['price']) #Subset - ustun nomini ko'rsatish
```

### Yuqoridagi misollarda bizning asl `df` jadvalimiz o'zgarishsiz qolmoqda, o'zgarishlarni `df` da saqlash uchun `inplace=True`parametridan foydalanami

```python
df.dropna(subset=['price'], inplace=True)
```
