```python
import sqlite3 as sql
import pandas as pd
import numpy as np
```

```python
db = sql.connect('new.db')
cursor = db.cursor()
```

```python
queary = """
CREATE TABLE IF NOT EXISTS newuu(
id SERIAL PRIMARY KEY,
country VARCHAR(255),
population INT,
area INT
);
"""
cursor.execute(queary)
db.commit()
```

```python
queary = "PRAGMA table_info(newuu);"
cursor.execute(queary)
cursor.fetchall()
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv")
df.head()
```

```python
avto = df[df.model == ' G Class']
```

```python
id = avto.index.values[0]
country = avto.at[id, 'model']
population = avto.at[id, 'mileage']
area = avto.at[id, 'tax']
print(f"{country=}, {population=}, {area=}")
```

```python
queary = f"""
INSERT INTO newuu (id, country, population, area) VALUES ({id},'{country}',{population}, {area});
"""
cursor.execute(queary)
db.commit()
```
