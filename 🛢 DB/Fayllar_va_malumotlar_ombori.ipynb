# `pandas` kutubxonasi fayllar bilan ishlaydigan bir nechta funksiyalardan iboraat

`read_csv` - vergul bilan ajratilgan jadval fayllarni o'qish uchun

`read_table` - probel(bo'sh joy) bilan ajratilgan fayllarni o'qish uchun

`read_excel` - `xls` yoki `xlsx` formatdagi Excel fayylarni o'qish uchun

`read_clipboard` - ko'chirilgan(xotirdagi) jadvallarni o'qish uchun

`read_hdf` - `HDF5` formatdagi fayllarni o'qish uchun

`read_html` - HTML (web) sahifadagi barcha jadvallarni o'qish

`read_json` - `JSON` ma'lumotlarni o'qish

`read_pickle` - pickle formatdagi fayllarni o'qish

`read_sql` - SQL query dan qaytgan ma'lumotlarni o'qish
