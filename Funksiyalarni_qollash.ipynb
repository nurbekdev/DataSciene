# Funksiyalarni qo'llash

```python
import pandas as pd
import numpy as np
```

Obyektlarni elemtlariga funksiya qo'llash

```python
df = pd.DataFrame(np.random.randn(3, 4), index=list('abc'), columns=['Olma', 'Anor', 'Qovun', 'Anjir'])
df
```

```python
np.round(df) #Soning butun qismini qaytaradi
```

```python
np.abs(df) # Soni musbat qilib beradi
```

```python
df.mean() # O'rtacha qiymat ustunlari uchun
```

```python
df.mean(axis=1) # Qatorlar uchun o'rtacha qiymat
```

```python
df.apply(np.abs) # Funksiya qo'llash, buni ustunligi o'zimizni funksiyalarimizni qo'llashimiz mumkin
```

```python
df.apply(sum, axis=1)
```

```python
f = lambda x: x.max() - x.min()
```

```python
df.apply(f)
```

```python
def foo(x):
  return pd.Series([x.min(), x.max()], index=['min', 'max'])
```

```python
df.apply(foo)
```

```python
df.applymap(lambda x: x*x) # Har bir element uchun bitta funksiyani qo'llash uchun ishlatiladi
```
